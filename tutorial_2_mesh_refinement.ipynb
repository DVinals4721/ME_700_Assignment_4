```python
from dolfinx import mesh, fem, log, default_scalar_type
from dolfinx.fem.petsc import NonlinearProblem
from dolfinx.nls.petsc import NewtonSolver
import numpy as np
import ufl
import matplotlib.pyplot as plt  # Import matplotlib without setting backend
import warnings
from mpi4py import MPI

# Add this line for Jupyter notebook inline plotting
%matplotlib inline  

print("Cantilever Beam FEA: h-refinement and p-refinement Study with FEniCSx")

# Beam geometry
L = 10.0   # length in x
H = 1.0    # height in y

# Material properties
E = 1000.0
nu = 0.3
mu = E / (2.0 * (1.0 + nu))
kappa = E / (3.0 * (1.0 - 2.0 * nu))
lmbda = E * nu / ((1.0 + nu) * (1.0 - 2.0 * nu))

print(f"Material properties: mu={mu:.3f}, kappa={kappa:.3f}")

# Load (large load for nonlinear behavior)
q = -0.5  # load per unit length in y

# Analytical solution (note: this will not be accurate for large deformations)
E_eff = E / (1 - nu ** 2.0)
I = H ** 3 / 12.0
w_analytical = q * L ** 4 / (8.0 * E_eff * I)
print(f"Analytical solution (small deformation theory): {w_analytical:.6f}")

def run_analysis(nx, ny, polynomial_order):
    """
    Run FEA for the cantilever beam problem with specified mesh density and polynomial order
    """
    print(f"  Mesh: {nx}x{ny} elements, polynomial order: {polynomial_order}")
    
    # Create mesh
    domain = mesh.create_rectangle(
        MPI.COMM_WORLD, 
        points=[[0.0, 0.0], [L, H]],
        n=[nx, ny],
        cell_type=mesh.CellType.triangle
    )
    
    # Define function space with specified polynomial order
    V = fem.functionspace(domain, ("Lagrange", polynomial_order, (domain.geometry.dim,)))
    
    # Define boundary conditions
    def left_boundary(x): return np.isclose(x[0], 0.0)
    def top_boundary(x): return np.isclose(x[1], H)
    
    # Mark boundaries
    fdim = domain.topology.dim - 1
    left_facets = mesh.locate_entities_boundary(domain, fdim, left_boundary)
    top_facets = mesh.locate_entities_boundary(domain, fdim, top_boundary)
    
    # Create facet tags
    marked_facets = np.hstack([left_facets, top_facets])
    marked_values = np.hstack([
        np.full_like(left_facets, 1),  # left boundary = 1
        np.full_like(top_facets, 2)    # top boundary = 2
    ])
    sorted_indices = np.argsort(marked_facets)
    facet_tag = mesh.meshtags(
        domain, fdim, 
        marked_facets[sorted_indices], 
        marked_values[sorted_indices]
    )
    
    # Define fixed left boundary condition
    u_bc = np.array((0, 0), dtype=default_scalar_type)
    left_dofs = fem.locate_dofs_topological(V, fdim, facet_tag.find(1))
    bcs = [fem.dirichletbc(u_bc, left_dofs, V)]
    
    # Define solution and test functions
    u = fem.Function(V, name="Displacement")
    v = ufl.TestFunction(V)
    
    # Neo-Hookean hyperelastic model
    d = len(u)
    I = ufl.variable(ufl.Identity(d))
    F = ufl.variable(I + ufl.grad(u))      # Deformation gradient
    C = ufl.variable(F.T * F)              # Right Cauchy-Green tensor
    Ic = ufl.variable(ufl.tr(C))           # First invariant of C
    J = ufl.variable(ufl.det(F))           # Volume ratio
    
    # Stored energy density (Neo-Hookean)
    psi = (mu / 2) * (Ic - d) - mu * ufl.ln(J) + (lmbda / 2) * (ufl.ln(J))**2
    
    # First Piola-Kirchhoff stress tensor
    P = ufl.diff(psi, F)
    
    # Define distributed load on top edge
    t = fem.Constant(domain, default_scalar_type((0, 0)))
    
    # Define measures for integration
    dx = ufl.Measure("dx", domain=domain)
    ds = ufl.Measure("ds", domain=domain, subdomain_data=facet_tag)
    
    # Variational form (internal - external forces)
    F_internal = ufl.inner(P, ufl.grad(v)) * dx
    F_external = ufl.inner(t, v) * ds(2)   # Applied on top edge (tag 2)
    F_form = F_internal - F_external
    
    # Set up nonlinear solver
    problem = NonlinearProblem(F_form, u, bcs)
    solver = NewtonSolver(MPI.COMM_WORLD, problem)
    solver.convergence_criterion = "incremental"
    solver.rtol = 1e-10
    solver.atol = 1e-10
    solver.max_it = 30
    
    # Apply load incrementally for better convergence
    num_steps = 20
    
    for step in range(1, num_steps + 1):
        # Update load gradually
        t.value = (0, q * step / num_steps)
        
        print(f"    Step {step}/{num_steps}: load = {t.value[1]:.3f}")
        
        # Solve the problem
        try:
            num_its, converged = solver.solve(u)
            if not converged:
                print(f"    WARNING: Newton solver did not converge at step {step}")
            u.x.scatter_forward()
            print(f"    Converged in {num_its} iterations")
        except Exception as e:
            print(f"    ERROR at step {step}: {e}")
            break
    
    # Find the tip node (middle of right edge)
    def find_tip_node(coords):
        """Find the node closest to the middle of the right edge"""
        right_edge_tol = 1e-10
        right_edge_nodes = np.where(np.isclose(coords[:, 0], L, atol=right_edge_tol))[0]
        if len(right_edge_nodes) == 0:
            return None
            
        # Find the node closest to the middle height
        middle_height = H / 2
        distances = np.abs(coords[right_edge_nodes, 1] - middle_height)
        closest_idx = np.argmin(distances)
        return right_edge_nodes[closest_idx]
    
    # Get coordinates of all nodes
    coords = domain.geometry.x
    
    # Find tip node
    tip_node = find_tip_node(coords)
    if tip_node is None:
        raise ValueError("Could not find tip node at x=L")
    
    # Get displacement at tip node
    u_values = u.x.array.reshape(-1, 2)
    tip_disp_y = u_values[tip_node, 1]  # y-component
    
    # Calculate mesh quality metrics
    mesh_quality = calculate_mesh_quality(domain)
    
    print(f"  Tip displacement: {tip_disp_y:.6f}")
    
    return tip_disp_y, mesh_quality

def calculate_mesh_quality(domain):
    """Calculate basic mesh quality metrics"""
    # Get mesh geometry
    coordinates = domain.geometry.x
    
    # Get mesh connectivity for cells
    tdim = domain.topology.dim
    domain.topology.create_connectivity(tdim, 0)
    connectivity = domain.topology.connectivity(tdim, 0)
    
    # Initialize arrays to store quality metrics
    num_cells = domain.topology.index_map(tdim).size_local
    aspect_ratios = np.zeros(num_cells)
    min_angles = np.zeros(num_cells)
    max_angles = np.zeros(num_cells)
    
    # Compute metrics for each cell
    for cell_idx in range(num_cells):
        # Get node indices for this cell
        nodes = connectivity.links(cell_idx)
        
        # Get node coordinates
        cell_coords = coordinates[nodes]
        
        # For triangular elements
        if len(nodes) == 3:
            # Compute edge lengths
            edges = np.zeros(3)
            edges[0] = np.linalg.norm(cell_coords[1] - cell_coords[0])  # edge 0-1
            edges[1] = np.linalg.norm(cell_coords[2] - cell_coords[1])  # edge 1-2
            edges[2] = np.linalg.norm(cell_coords[0] - cell_coords[2])  # edge 2-0
            
            # Compute aspect ratio
            if np.min(edges) > 0:
                aspect_ratios[cell_idx] = np.max(edges) / np.min(edges)
            else:
                aspect_ratios[cell_idx] = float('inf')
            
            # Compute angles using law of cosines
            angles = np.zeros(3)
            
            # Angle at vertex 0
            if edges[0] * edges[2] > 0:
                cos_angle = ((cell_coords[1] - cell_coords[0]) @ (cell_coords[2] - cell_coords[0])) / (edges[0] * edges[2])
                angles[0] = np.arccos(np.clip(cos_angle, -1.0, 1.0))
            
            # Angle at vertex 1
            if edges[0] * edges[1] > 0:
                cos_angle = ((cell_coords[0] - cell_coords[1]) @ (cell_coords[2] - cell_coords[1])) / (edges[0] * edges[1])
                angles[1] = np.arccos(np.clip(cos_angle, -1.0, 1.0))
            
            # Angle at vertex 2
            if edges[1] * edges[2] > 0:
                cos_angle = ((cell_coords[1] - cell_coords[2]) @ (cell_coords[0] - cell_coords[2])) / (edges[1] * edges[2])
                angles[2] = np.arccos(np.clip(cos_angle, -1.0, 1.0))
            
            # Convert to degrees
            angles = np.degrees(angles)
            min_angles[cell_idx] = np.min(angles)
            max_angles[cell_idx] = np.max(angles)
    
    return {
        'aspect_ratios': aspect_ratios,
        'min_angles': min_angles,
        'max_angles': max_angles
    }

def visualize_mesh_quality(mesh_quality, title, filename):
    """Create histograms of mesh quality metrics with robust bin calculation"""
    fig, axs = plt.subplots(1, 3, figsize=(15, 5))
    
    # Helper function to safely create histograms
    def safe_hist(ax, data, title_str, xlabel_str):
        try:
            # Let matplotlib automatically determine the number of bins
            ax.hist(data, bins='auto')
            ax.set_title(title_str)
            ax.set_xlabel(xlabel_str)
            ax.set_ylabel('Count')
        except ValueError as e:
            warnings.warn(f"Could not create histogram: {e}")
            ax.text(0.5, 0.5, f"Error creating histogram:\n{str(e)}", 
                    ha='center', va='center', transform=ax.transAxes)
            ax.set_title(title_str)
    
    # Create histograms with safe function
    safe_hist(axs[0], mesh_quality['aspect_ratios'], 'Aspect Ratio', 'Aspect Ratio')
    safe_hist(axs[1], mesh_quality['min_angles'], 'Minimum Angle', 'Angle (degrees)')
    safe_hist(axs[2], mesh_quality['max_angles'], 'Maximum Angle', 'Angle (degrees)')
    
    fig.suptitle(title)
    plt.tight_layout()
    plt.show()  # Show directly in notebook

# ---------------------------------------------------------
# h-refinement study
# ---------------------------------------------------------
print("\n----- h-refinement study -----")
h_refinement_factors = [2, 5, 10, 15, 20]
h_refinement_results = []
h_refinement_errors = []
h_refinement_qualities = []

for n in h_refinement_factors:
    print(f"\nAnalyzing mesh with {n} elements in x-direction")
    tip_disp, mesh_quality = run_analysis(n, 2, 2)  # Using quadratic elements
    h_refinement_results.append(tip_disp)
    h_refinement_errors.append(abs(tip_disp - w_analytical))
    h_refinement_qualities.append(mesh_quality)
    
    # Create element quality histograms
    visualize_mesh_quality(
        mesh_quality,
        f"Mesh Quality Metrics (h-refinement, n={n})",
        f"h_refinement_quality_{n}.png"
    )

# Plot h-refinement results
plt.figure(figsize=(10, 6))
plt.plot(h_refinement_factors, h_refinement_results, marker='o')
plt.axhline(y=w_analytical, color='r', linestyle='--', label='Analytical solution')
plt.xlabel("Number of elements in x-direction")
plt.ylabel("Tip displacement")
plt.title("h-refinement study")
plt.legend()
plt.grid(True)
plt.show()  # Show directly in notebook

# Plot h-refinement errors
plt.figure(figsize=(10, 6))
plt.plot(h_refinement_factors, h_refinement_errors, marker='o')
plt.xlabel("Number of elements in x-direction")
plt.ylabel("Absolute error")
plt.title("h-refinement error")
plt.yscale('log')
plt.grid(True)
plt.show()  # Show directly in notebook

# ---------------------------------------------------------
# p-refinement study
# ---------------------------------------------------------
print("\n----- p-refinement study -----")
polynomial_orders = [1, 2, 3, 4]  # Linear, quadratic, cubic, quartic elements
element_type_names = ["Linear", "Quadratic", "Cubic", "Quartic"]
p_refinement_results = []
p_refinement_errors = []
p_refinement_qualities = []

for p in polynomial_orders:
    print(f"\nAnalyzing with polynomial order: {p} ({element_type_names[p-1]} elements)")
    tip_disp, mesh_quality = run_analysis(20, 2, p)
    p_refinement_results.append(tip_disp)
    p_refinement_errors.append(abs(tip_disp - w_analytical))
    p_refinement_qualities.append(mesh_quality)
    
    # Create element quality histograms
    visualize_mesh_quality(
        mesh_quality,
        f"Mesh Quality Metrics (p-refinement, order={p})",
        f"p_refinement_quality_{p}.png"
    )

# Plot p-refinement results
plt.figure(figsize=(10, 6))
plt.plot(element_type_names, p_refinement_results, marker='o')
plt.axhline(y=w_analytical, color='r', linestyle='--', label='Analytical solution')
plt.xlabel("Element type")
plt.ylabel("Tip displacement")
plt.title("p-refinement study")
plt.legend()
plt.grid(True)
plt.show()  # Show directly in notebook

# Plot p-refinement errors
plt.figure(figsize=(10, 6))
plt.plot(element_type_names, p_refinement_errors, marker='o')
plt.xlabel("Element type")
plt.ylabel("Absolute error")
plt.title("p-refinement error")
plt.yscale('log')
plt.grid(True)
plt.show()  # Show directly in notebook

# ---------------------------------------------------------
# Display summary of results
# ---------------------------------------------------------
print("\n----- Results Summary -----")
print(f"Analytical solution (small deformation): {w_analytical:.6f}")

print("\nh-refinement results:")
for n, result, error in zip(h_refinement_factors, h_refinement_results, h_refinement_errors):
    print(f"nx = {n}: tip displacement = {result:.6f}, error = {error:.6e}")

print("\np-refinement results:")
for p, name, result, error in zip(polynomial_orders, element_type_names, p_refinement_results, p_refinement_errors):
    print(f"{name} (order = {p}): tip displacement = {result:.6f}, error = {error:.6e}")
```
